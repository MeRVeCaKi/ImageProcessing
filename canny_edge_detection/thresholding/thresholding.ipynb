```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import display, Image
from PIL import Image as pimg
```

```python
image_path = "sharpened_lena.png"
image = cv2.imread(image_path)
print(image.shape)
```

```python
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
print(image.shape)
```

```python
height , width = image.shape[0], image.shape[1]
```

```python
unique_pixels = []
pixels = []
for y in range(0, height):
    for x in range(0, width):
        pixels.append(image[y,x])
        if image[y,x] not in unique_pixels:
            unique_pixels.append(image[y,x])
print(unique_pixels)
print(pixels)
```

```python
unique_pixels.sort()
print(unique_pixels)
```

```python
histogram = {}
for pixel_value in unique_pixels:
    histogram[pixel_value] = pixels.count(pixel_value)
print(histogram)
```

```python
plt.bar(histogram.keys(), histogram.values(), color='gray')
plt.grid(True)
plt.show()
```

```python
del histogram[0]
```

```python
for y in range(0, height):
    for x in range(0, width):
        if(image[y,x] > 127):
            image[y,x] = 255
        else:
            image[y,x] = 0
```

```python
cv2.imwrite('thresholding.png', image.astype(np.uint8))
display(Image(filename='thresholding.png'))
```

```python
image1 = pimg.open('sharpened_lena.png')
image2 = pimg.open('thresholding.png')
combined_image = pimg.new('RGB', (image1.width + image2.width, image1.height))
combined_image.paste(image1, (0, 0))
combined_image.paste(image2, (image1.width, 0))
display(combined_image)
```
